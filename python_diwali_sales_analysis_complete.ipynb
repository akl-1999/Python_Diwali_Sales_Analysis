```python
#Reading The CSV File
import pandas as pd
df=pd.read_csv("diwalisalesdata.csv",encoding='unicode_escape')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df['Amount']=df["Amount"].astype(int)
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df.describe()
```

```python
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

ax=sns.histplot(data=df,x="Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby('Gender')['Amount'].sum()
```

```python
sales_data=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.barplot(data=sales_data,x='Gender',y='Amount')
plt.show()
```

from the above graphs we can see that most of the byers are female even the purchase power of females are grater then male

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.show()    
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.barplot(data=sales_age,x='Age Group',y='Amount')
plt.show()
```

from above graphs we can see that most of the buyers are from age group 26-35 and female

```python
sales_age=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values('Orders',ascending=False).head(10)
sns.barplot(data=sales_age,x='State',y='Orders')
sns.set_theme(rc={'figure.figsize':(20,5)})
plt.show()
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False).head(10)
sns.barplot(data=sales_state,x='State',y='Amount')
sns.set_theme(rc={'figure.figsize':(15,5)})
plt.show()
```

from above graphs we can see that most of the orders are from uttar prades,maharastra,karnataka

```python
ax=sns.countplot(data=df,x='Marital_Status')

for bars in ax.containers:
    ax.bar_label(bars)
    
plt.show()
```

```python
sales_marital=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.barplot(data=sales_marital,x='Marital_Status',y='Amount',hue='Gender')
sns.set_theme(rc={'figure.figsize':(7,5)})
plt.show()
```

from above graphs we can see that most buyers are from married(women) and with higher purchase power

```python


ax=sns.histplot(data=df,x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
sns.set_theme(rc={'figure.figsize':(28,5)})    
plt.show()
```

```python
sales_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.barplot(data=sales_occupation,x='Occupation',y='Amount')
sns.set_theme(rc={'figure.figsize':(7,5)})
plt.show()
```

from above graphs we can see that buyers are from It, healthcare and aviation sector

```python
ax=sns.histplot(data=df,x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
sns.set_theme(rc={'figure.figsize':(50,5)})    
plt.show()
```

```python
sales_product=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values('Amount',ascending=False)
sns.barplot(data=sales_product,x='Product_Category',y='Amount')
sns.set_theme(rc={'figure.figsize':(35,7)})
plt.show()
```

from the abpove graphs we can see that most sold product category are food,clothing and electronics
